```python
import glob

project_files = glob.glob("../data/projects/*.json")

print("Length of project_files: ", len(project_files))
```

```python
import json
import os

for project in project_files:
    print("Process: ", project)

    with open(project, 'r') as f:
        data = json.load(f)

    output_file = f"../data/projects_last_commit/{data['project_name']}_last_commit.json"
    if os.path.exists(output_file):
        print(f"File {output_file} already exists, skipping.")
        continue
        
    latest_commit = data["commit_data"][-1]

    del data["commit_data"]

    data["latest_commit_data"] = latest_commit

    with open(output_file, 'w') as f:
        json.dump(data, f, indent=4)
```
