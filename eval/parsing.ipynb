**Spring Boot**

```python
import requests
from bs4 import BeautifulSoup


url = "https://docs.spring.io/spring-boot/appendix/application-properties/index.html"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

output_file = "../data/technologies/spring_boot.properties"
properties = []

# Iterate over all tables containing properties
for table in soup.select('tbody'):
    for row in table.select('tr'):
        columns = row.find_all('td')
        if len(columns) >= 2:
            option = columns[0].get_text(strip=True)
            default = columns[2].get_text(strip=True)
            # Skip entries that are empty or malformed
            if option:
                properties.append((option, default))


with open(output_file, 'w', encoding='utf-8') as f:
    for option, default in properties:
        sanitized_default = default.replace('\u00A0', ' ').strip()
        f.write(f"{option}={sanitized_default}\n" if sanitized_default else f"{option}=\n")
```

**MySQL**

```python
import requests
from bs4 import BeautifulSoup


url = "https://dev.mysql.com/doc/refman/8.4/en/server-system-variables.html"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

output_file = "../data/technologies/mysql.properties"
properties = []

properties = []
for list in soup.select("ul"):  # Skip header
    for li in list.select("li"):
        option = li.find("code", class_="literal")
        if option:
            option_name = option.get_text(strip=True)
            default_value= ""

            table = li.find("table")
            if table:
                for row in table.select("tr"):
                    th = row.find("th")
                    td = row.find("td")
                    if th and td and th.get_text(strip=True) == "Default Value":
                        code = td.find("code", class_="literal")
                        default_value = code.get_text(strip=True) if code else td.get_text(strip=True)

                properties.append((option_name, default_value))

with open(output_file, 'w', encoding='utf-8') as f:
    for option, default in properties:
        f.write(f"{option}={default}\n")
```

```python
import requests, re
from bs4 import BeautifulSoup
from urllib.parse import urljoin

INDEX_URL = "https://kubernetes.io/docs/reference/config-api/"

def get_config_api_urls():
    r = requests.get(INDEX_URL)
    soup = BeautifulSoup(r.text, "html.parser")
    return { urljoin(INDEX_URL, a['href'])
             for a in soup.select("ul li a[href*='config-api/']") }

def extract_fields(url):
    print("Parsing", url)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    properties = []
    table_rows = soup.find_all("tr")
    for row in table_rows:
        cells = row.find_all("td")
        if cells:
            option_cell = cells[0]
            option_text = option_cell.find("code").get_text(strip=True)
            if option_text:
                properties.append((option_text, ""))

    return properties


urls = get_config_api_urls()
all_properties = []
for u in urls:
    all_properties.extend(extract_fields(u))

with open("../data/technologies/kubernetes.properties", 'w', encoding='utf-8') as f:
    for option, default in set(all_properties):
        f.write(f"{option}={default}\n")
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import time

BASE = "https://httpd.apache.org/docs/current/mod/directives.html"

def get_directive_links():
    res = requests.get(BASE)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, "html.parser")

    directive_links = []
    directive_list = soup.find("div", {"id": "directive-list"})
    for element in directive_list.find_all("a", href=True):
        href = element["href"]
        full_url = urljoin("https://httpd.apache.org/docs/current/mod/", href)
        directive_links.append(full_url)

    return sorted(set(directive_links))


def extract_default(url):
    res = requests.get(url)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, "html.parser")
    properties = []

    # Get directive name from <h1>
    sections = soup.find_all("div", class_="directive-section")
    for section in sections:
        option_header= section.find("h2")
        option_name = option_header.find("a").get_text(strip=True)
        if option_name.startswith("<") and option_name.endswith(">"):
            option_name = option_name[1:-1].strip()

        default_value = ""
        table = section.find("table", class_="directive")
        for row in table.find_all("tr"):
            header = row.find("th")
            if "Default" in header.get_text():
                cell = row.find("td")
                if cell:
                    parts = cell.get_text(strip=True).split(" ")  # 
                    default_value = " ".join(parts[1:])  # Join parts to handle multi-word defaults

        properties.append((option_name, default_value))

    
    return properties


links = get_directive_links()
print(f"Found {len(links)} directive pages.")

all_properties = []
for url in links:
    print(url)
    all_properties.extend(extract_default(url))

with open("../data/technologies/apache_webserver.properties", "w", encoding="utf-8") as f:
    for option, default in set(all_properties):
        f.write(f"{option}={default}\n")
```

```python
import requests
from bs4 import BeautifulSoup
import re

url = "https://docs.flutter.dev/tools/pubspec"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

output_file = "../data/technologies/flutter.properties"
properties = []

# Find all h3 headings like "assets field", "config field", ...
for heading in soup.find_all("h3"):
    heading_text = heading.get_text(strip=True)

    match = re.match(r"(.+?) field$", heading_text, re.IGNORECASE)
    if match:
        option = match.group(1)  # e.g. "assets", "config", "default-flavor", ...
        default_value = ""
        properties.append((option, default_value))

with open(output_file, "w", encoding="utf-8") as f:
    for option, default in properties:
        f.write(f"{option}={default}\n")
```

**Hadoop Hbase**

```python
import requests
import xml.etree.ElementTree as ET

# hbase-default.xml used to generate the "Default Configuration" section in the docs
url = "https://raw.githubusercontent.com/apache/hbase/master/hbase-common/src/main/resources/hbase-default.xml"

response = requests.get(url)
response.raise_for_status()

xml_content = response.content

output_file = "../data/technologies/hadoop_hbase.properties"
properties = []

# Parse the XML
root = ET.fromstring(xml_content)

# hbase-default.xml is a <configuration> with many <property> children
for prop in root.findall(".//property"):
    name_el = prop.find("name")
    value_el = prop.find("value")

    if name_el is None or name_el.text is None:
        continue

    option = name_el.text.strip()
    default_value = ""

    if value_el is not None and value_el.text is not None:
        default_value = value_el.text.strip().replace("\n", " ").replace("\r", " ")

    properties.append((option, default_value))

# Write out as key=value, like your RabbitMQ example
with open(output_file, "w", encoding="utf-8") as f:
    for option, default in properties:
        # Very light escaping so the file stays usable as a .properties file
        default = default.replace("\\", "\\\\").replace("=", r"\=").replace(":", r"\:")
        f.write(f"{option}={default}\n")
```

**Kafka**

```python


import requests
from bs4 import BeautifulSoup
import re

urls = [
    "https://docs.redhat.com/de/documentation/red_hat_streams_for_apache_kafka/2.4/html/kafka_configuration_properties/broker-configuration-properties-str",
    "https://docs.redhat.com/de/documentation/red_hat_streams_for_apache_kafka/2.4/html/kafka_configuration_properties/topic-configuration-properties-str",
    "https://docs.redhat.com/de/documentation/red_hat_streams_for_apache_kafka/2.4/html/kafka_configuration_properties/consumer-configuration-properties-str",
    "https://docs.redhat.com/de/documentation/red_hat_streams_for_apache_kafka/2.4/html/kafka_configuration_properties/producer-configuration-properties-str",
    "https://docs.redhat.com/de/documentation/red_hat_streams_for_apache_kafka/2.4/html/kafka_configuration_properties/admin-client-configuration-properties-str",
    "https://docs.redhat.com/de/documentation/red_hat_streams_for_apache_kafka/2.4/html/kafka_configuration_properties/kafka-connect-configuration-properties-str",
    "https://docs.redhat.com/de/documentation/red_hat_streams_for_apache_kafka/2.4/html/kafka_configuration_properties/kafka-streams-configuration-properties-str"
]

output_file = "../data/technologies/kafka.properties"
properties = []

for url in urls:
    print("Parse: ", url)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    dl_block = soup.find("dl", class_="variablelist")

    dt_blocks = dl_block.find_all("dt")
    dd_blocks = dl_block.find_all("dd")

    for dt, dd in zip(dt_blocks, dd_blocks):
        option = dt.get_text().strip()
        default_value = ""
        spans = dd.find_all("span")
        for span in spans:
            strong = span.find("strong")
            if strong and strong.get_text(strip=True).lower().startswith("default:"):
                # The default value is in the *next sibling text node*
                next_text = span.next_sibling

                if next_text:
                    default_value = next_text.strip().strip('"').split(" ")[0]
                break
        
        properties.append((option, default_value))


with open(output_file, 'w', encoding='utf-8') as f:
    for option, default in properties:
       f.write(f"{option}={default}\n")
```

**PostgreSQL**

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import re

base_url = "https://www.postgresql.org/docs/current/runtime-config.html"
output_file = "../data/technologies/postgresql.properties"

properties = set()

# 1. Collect all runtime-config* pages (overview + subsections)
response = requests.get(base_url)
response.raise_for_status()
soup = BeautifulSoup(response.content, "html.parser")

page_urls = {base_url}

for a in soup.find_all("a", href=True):
    href = a["href"]
    if href.startswith("runtime-config-") and href.endswith(".html"):
        full = urljoin(base_url, href)
        page_urls.add(full)

for url in sorted(page_urls):
    resp = requests.get(url)
    resp.raise_for_status()
    page_soup = BeautifulSoup(resp.content, "html.parser")

    var_lists = page_soup.find_all("div", class_="variablelist")

    for var_list in var_lists:
        dt_blocks = var_list.find_all("dt")
        for dt in dt_blocks:
            option_name = dt.get_text().split(" ")[0]
            default_value = ""

            properties.add((option_name, default_value))

with open(output_file, "w", encoding="utf-8") as f:
    for option, default in sorted(properties, key=lambda x: x[0]):
        f.write(f"{option}={default}\n")
```

**Redis**

```python
import requests

# Self-documented redis.conf (here: Redis 7.4)
url = "https://raw.githubusercontent.com/redis/redis/7.4/redis.conf"
response = requests.get(url)
response.raise_for_status()

output_file = "../data/technologies/redis.properties"

options = []

seen = set()

for line in response.text.splitlines():
    stripped = line.strip()

    # Skip empty lines and comments
    if stripped or not stripped.startswith("#"):
        parts = stripped.split()

        if not parts:
            continue

        directive = parts[0]

        if directive not in seen:
            seen.add(directive)
            options.append(directive)

with open(output_file, "w", encoding="utf-8") as f:
    for option in sorted(options):
        default_value = ""
        f.write(f"{option}={default_value}\n")
```
